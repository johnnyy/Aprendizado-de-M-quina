```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
time_step = 1
period = 5.
time_vec = np.arange(0, 5000, time_step)
sig = np.sin(2 * np.pi / period * time_vec) + \
             0.5 * np.random.randn(time_vec.size)
```

```python
from scipy import fftpack
sample_freq = fftpack.fftfreq(sig.size, d=time_step)
sig_fft = fftpack.fft(sig)
```

```python
pidxs = np.where(sample_freq > 0)
freqs = sample_freq[pidxs]
power = np.abs(sig_fft)[pidxs]
```

```python
pl.figure()
pl.plot(freqs, power)
pl.xlabel('Frequencia [Hz]')
pl.ylabel('Energia')
axes = pl.axes([0.3, 0.3, 0.5, 0.5])
pl.title('Pico de frequencia')
pl.plot(freqs[:8], power[:8])
pl.setp(axes, yticks=[])
```

```python
freq = freqs[power.argmax()]
np.allclose(freq, 1./period)  # checa se aquele frequência correta é encontrada
```

```python
sig_fft[np.abs(sample_freq) > freq] = 0
```

```python
main_sig = fftpack.ifft(sig_fft)
```

```python
import pylab as plt

plt.figure()
plt.plot(time_vec, sig)
plt.plot(time_vec, main_sig, linewidth=3)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
```

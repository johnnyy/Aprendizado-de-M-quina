```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
from sklearn import preprocessing
from sklearn import model_selection

from sklearn import linear_model
from sklearn import neighbors
from sklearn import svm
from sklearn import neighbors
from sklearn import naive_bayes
from sklearn import metrics
```

```python
data = pd.read_csv("Adult.csv",header=None)
```

## Substituindo valores 0 pela média entre os valores das colunas 10 e 11

```python
data_M = data.iloc[:]
Imp = preprocessing.Imputer(missing_values=0,axis=0,strategy='mean')
Imp.fit(data.iloc[:,[10,11]])
data_M.loc[:,[10,11]] = Imp.transform(data.iloc[:,[10,11]])

data_M.head()
```

```python
for i in [1,3,5,6,7,8,9,13,14]:
    Enc =  preprocessing.LabelEncoder()
    data_M.iloc[:,i] = Enc.fit_transform(data_M.iloc[:,i])
    


ohe = preprocessing.OneHotEncoder(categorical_features=[1,3,5,6,7,8,9,13])
data_Moral = ohe.fit_transform(data_M).toarray()
X = data_Moral[:,:108]
y = data_Moral[:,108]
```

## Na matriz mostrada abaixo é informado os valores antes de usar o OneHotEnconder, gerando uma tabela de 100 colunas.

```python
data_M
```

```python

fold =  model_selection.StratifiedKFold(n_splits=5)
mediaKNN = 0
mediaSVM = 0
mediaLR = 0
mediaNB = 0

for train,test in fold.split(X,y):
    
    
    Stand = preprocessing.StandardScaler()
    Stand.fit(X[train])
    X_train = Stand.transform(X[train])
    X_test =  Stand.transform(X[test])
    
    
    modelK = neighbors.KNeighborsClassifier()
    modelSVM = svm.SVC()
    modelLR = linear_model.LogisticRegression()
    modelNB = naive_bayes.GaussianNB()
    
    
    modelK.fit(X_train,y[train])
    modelSVM.fit(X_train,y[train])
    modelLR.fit(X_train,y[train])
    modelNB.fit(X_train,y[train])
    
    
    y_predK = modelK.predict(X_test)
    y_predSVM = modelSVM.predict(X_test)
    y_predLR = modelLR.predict(X_test)
    y_predNB = modelNB.predict(X_test)
    
    
    accK = metrics.accuracy_score(y[test],y_predK)
    mediaKNN = mediaKNN + accK
    accSVM = metrics.accuracy_score(y[test],y_predSVM)
    mediaSVM =mediaSVM+ accSVM
    accLR = metrics.accuracy_score(y[test],y_predLR)
    mediaLR =mediaLR+ accLR
    accNB = metrics.accuracy_score(y[test],y_predNB)
    mediaNB =mediaNB+ accNB
    
    print("KNN",accK)
    print("SVM",accSVM)
    print("LR",accLR)
    print("NB",accNB)
    
    print()
mediaKNN = mediaKNN/5
mediaSVM = mediaSVM/5
mediaLR = mediaLR/5
mediaNB = mediaNB/5


print("Media KNN",mediaKNN)
print("Media SVM",mediaSVM)
print("Media LR",mediaLR)
print("Media NB",mediaNB)
```

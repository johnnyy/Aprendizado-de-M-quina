```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn import metrics
import numpy as np
from mylibs import metricas
```

```python
X = [[1],[2],[4],[3],[5]]
y = [1,3,3,2,5]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, y)
```

```python
plt.scatter(X,y)
```

```python
plt.scatter(X,y, marker='x')
plt.xlabel('Pontos X')
plt.ylabel('Pontos Y')
plt.grid()
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X,y)
```

# Coeficientes

```python
print('B0 = ', model.intercept_)
print('B1 = ', model.coef_)
```

## Usando o modelo para predizer as saidas a partir da entrada

```python
y_pred = model.predict(X)
print(y_pred)
print(y)
plt.grid()
plt.scatter(X,y, color='red')
plt.plot(X,y_pred)
```

## Avaliando o modelo linear

```python
mse1 = metrics.mean_squared_error(y,y_pred)
print(mse1)
```

4- Crie um modelo linear usando o Scikit Learn para representar os dados (todo o dataset).

```python
## B1 .. B10
model.coef_
```

```python
y_pred = model.predict(X[0:,0:])
```

```python
y_pred
```

```python
plt.grid()
plt.plot(X, y_pred, color = 'blue')
for i in range(0, 10):
    lista = X[0:,i]
    plt.scatter(lista, y, color = 'red')
```

5- Exiba a fórmula da função linear com todos os seus coeficientes.

```python
count = 1
result = "y = " + str(model.intercept_)
for i in model.coef_:
    if(i > 0):
        result += " + " + str(i) + " X[" + str(count) + "]"
    else:
        result += " " + str(i) + " X[" + str(count) + "]"
    count += 1
print(result)
```

8- Crie um novo modelo de regressão linear para a feature que melhor se ajusta a uma função linear (ver questão 2a).

```python
model2 = linear_model.LinearRegression()
lista_aux = [ [i] for i in X[0:,2] ]
model2.fit(lista_aux, y)
```

- Calcule e exiba os coeficientes da função linear de forma manual (usando o Python, mas sem usar o Scikit Learn). 

Para calcular $ \beta_{1} $ temos que:
$ \displaystyle \beta_1 = \frac{\sum_{i=1}^{n} (x_{i} - mean(x) \times (y_{i} - mean(y)))}{\sum_{i=1}^{n} (x_i - mean(x)^{2})} $

```python
y_ = [[i] for i in y]
B1 = (np.sum((lista_aux - np.mean(lista_aux)) * (y_ - np.mean(y_)))) / (np.sum((lista_aux - np.mean(lista_aux)) ** 2))
```

```python
B1
```

Para calcular $ \beta_{0} $ temos que:
$ \displaystyle \beta_0 = mean(y) - \beta_1 \times mean(x) $

```python
B0 = np.mean(y_) - B1 * np.mean(lista_aux)
```

```python
B0
```

```python
data = pd.read_csv("crime.csv")
data.head()
```

```python
names = ['0-Break and Enter Commercial', ' 1 -Break and Enter Residential/Other',
        ' 2 -Homicide', ' 3 -Mischief', ' 4 -Offence Against a Person', ' 5 -Other Theft',
        ' 6 -Theft from Vehicle', ' 7 -Theft of Bicycle', ' 8 -Theft of Vehicle',
        ' 9 -Vehicle Collision or Pedestrian Struck (with Fatality)',
        ' 10 -Vehicle Collision or Pedestrian Struck (with Injury)']
```

```python
data_Sta = data[data['Latitude'] != 0]

bloco = data_Sta.iloc[:,6].unique()
bairro = data_Sta.iloc[:,7].unique()
tipo = data_Sta.iloc[:,0].unique()

for i in range(tipo.shape[0]):
    data_Sta.iloc[:,0] = np.where(data_Sta.iloc[:,0] == tipo[i] ,i, data_Sta.iloc[:,0])

for i in range(bairro.shape[0]):
    data_Sta.iloc[:,7] = np.where(data_Sta.iloc[:,7] == bairro[i] ,i, data_Sta.iloc[:,7])
    
for i in range(bloco.shape[0]):
    data_Sta.iloc[:,6] = np.where(data_Sta.iloc[:,6] == bloco[i] ,i, data_Sta.iloc[:,6])

data_Sta.to_csv('crime_Sta.csv')    
data_Sta
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import calendar
import folium
import folium.plugins
from sklearn import linear_model
from sklearn import neighbors
from sklearn import svm
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import metrics
from sklearn import svm
from sklearn import naive_bayes
```

```python

dataX = data_Sta.iloc[:,1:]
datay = data_Sta.iloc[:,0]
dataX = dataX.dropna()
dataX.head()
dataX.
```

```python
Fold =  model_selection.KFold(n_splits=3)
```

```python

mediaaccKNN = 0
mediaaccL = 0
mediaaccS =0
for train,test in Fold.split(dataX):
    
    #KNN
    
    KNN =  neighbors.KNeighborsClassifier(n_neighbors=5,n_jobs=-1,weights='distance')
    KNN.fit(dataX.iloc[train],datay.iloc[train])
    y_predK = KNN.predict(dataX.iloc[test])
    accK = metrics.accuracy_score(datay.iloc[test], y_predK)
    mediaaccKNN = mediaaccKNN + accK
    print("Accuracy KNN:", accK)
    
    #Logistic Regression
    
    LR =  linear_model.LogisticRegression(n_jobs=-1)
    LR.fit(dataX.iloc[train],datay.iloc[train])
    y_predL = LR.predict(dataX.iloc[test])
    accL = metrics.accuracy_score(datay.iloc[test], y_predL)
    mediaaccL = mediaaccL + accL
    print("Accuracy LR:", accL)
    
    #SVM
    SVM = svm.SVC()
    SVM.fit(dataX.iloc[train],datay.iloc[train])
    y_predS = SVM.predict(dataX.iloc[test])
    accS = metrics.accuracy_score(datay.iloc[test], y_predS)
    mediaaccS = mediaaccS + accS
    print("Accuracy SVM:", accS)

    #KNN.predict(dataX)
mediaaccKNN = mediaaccKNN/Fold.n_splits
print("Média KNN:",mediaaccKNN)
mediaaccL = mediaaccL/Fold.n_splits
print("Média LR:",mediaaccL)
#mediaaccS = mediaaccS/Fold.n_splits
#print("Média SVM:",mediaaccS)
```

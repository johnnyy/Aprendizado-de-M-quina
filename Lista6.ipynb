## Questão 2

```python
data = pd.read_csv("winequality-white.csv", header=None)
```

```python
data.iloc[:,11] = data.iloc[:,11].astype(int)
data.head(9)
```

Separando Splits

```python
splits = 3
tx = range(data.shape[0])
ty = data.iloc[:,11]
train_index_1 = []
test_index_1 = []
kfold = StratifiedKFold(n_splits=splits, shuffle=True, random_state=42)
for train_index, test_index in kfold.split(tx,ty):
    print(train_index, test_index)
    train_index_1.append(train_index)
    test_index_1.append(test_index)
```

```python
testX = []
testy = []
trainX = []
trainy = []
for i in range(splits):
    testX.append(np.array(data.iloc[test_index_1[i],0:11]))
    testy.append(np.array(data.iloc[test_index_1[i],11]))
    trainX.append(np.array(data.iloc[train_index_1[i],:11]))
    trainy.append(np.array(data.iloc[train_index_1[i],11]))
```

```python
skl = KNeighborsClassifier(n_neighbors=3)
myModel1 = knn.KNNClassifier(k=3,metric='euclidean')
myModel2 = knn.KNNClassifier(k=3,metric='manhattan')
y_predsk = []
y_predM1 = []
y_predM2 = []
for i in range(splits):
    skl.fit(trainX[i],trainy[i])
    y_predsk.append(skl.predict(testX[i]))
    
    myModel1.fit(trainX[i],trainy[i])
    y_predM1.append(myModel1.predict(testX[i]))
    
    myModel2.fit(trainX[i],trainy[i])
    y_predM2.append(myModel2.predict(testX[i]))
```

```python
splits = 3
for i in range(splits):
    print("Split{}".format(i))
    print(np.unique(y_predsk[i],return_counts=True))
    print(np.unique(y_predM1[i],return_counts=True))
    print(np.unique(y_predM2[i],return_counts=True))
```
